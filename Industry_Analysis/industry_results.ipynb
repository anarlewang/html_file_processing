```python
import plotly.io as pio
pio.renderers
pio.renderers.default = "notebook_connected"

import pandas as pd 
```

```python

sic_category = ["Agriculture", "Mining", "Construction", "Manufacturing", 
                "Transportation", "Wholesale Trade", "Retail Trade", 
                "Finance, Insurance and Real Estate", "Services", "Public Administration"]

smaller = ["Finance, Insurance and Real Estate", "Services", "Manufacturing"]
cat = smaller[2]

df = pd.read_csv(cat+'_keywords_results.csv')
```

```python
df.head(3)
```

```python
df2 = df[["year", "comp_general", "comp_detailed", "dcf"]]
df3 = df2.groupby(["year"]).agg('sum')
df3['comp_all'] = df3['comp_general'] + df3['comp_detailed']
```

```python
df3.head(3)
```

```python



import plotly.express as px 
fig = px.bar(df3, x=df3.index, y=df3.columns[2:], title= cat)
fig.show()
```

```python
cat = smaller[0]
df = pd.read_csv(cat+'_keywords_results.csv')
df2 = df[["year", "comp_general", "comp_detailed", "dcf"]]
df3 = df2.groupby(["year"]).agg('sum')
df3['comp_all'] = df3['comp_general'] + df3['comp_detailed']


# fig = px.bar(df3, x=df3.index, y=df3.columns[2:], title= cat, color_discrete_map={'dcf': 'gold','comp_all': 'sandy brown'})
# fig.show()
```

```python
sub_cats = []

for i in df.sic_code: 
    if i <= 6099: 
        sub_cats.append('Depository Institutions')

    elif i > 6099 and i < 6200:
        sub_cats.append('Non-depository Credit Institutions')

    elif i >= 6200 and i < 6300:
        sub_cats.append('Security And Commodity Brokers, Dealers, Exchanges, And Services')
    
    elif i >= 6300 and i < 6400:
        sub_cats.append('Insurance Carriers')
    
    elif i >= 6400 and i < 6500:
        sub_cats.append('Insurance Agents, Brokers, And Service')
    
    elif i >= 6500 and i < 6600:
        sub_cats.append('Real Estate')
    
    elif i >= 6700 and i < 6800:
        sub_cats.append('Holding And Other Investment Offices')
    

```

```python
df['sub_sic'] = sub_cats
df.head(4)
```

```python
df2 = df[["comp_general", "comp_detailed", "dcf", "sub_sic"]]
df3 = df2.groupby(["sub_sic"]).agg('sum')
df3['comp_all'] = df3['comp_general'] + df3['comp_detailed']
```

```python
fig = px.bar(df3, x=df3.index, y=df3.columns[2:], title= cat, color_discrete_map={'dcf': 'gold','comp_all': 'sandy brown'})
fig.show()
```
